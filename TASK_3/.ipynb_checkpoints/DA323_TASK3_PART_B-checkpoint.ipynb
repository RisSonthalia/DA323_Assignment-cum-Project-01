
# Visual Analysis of National Flag Images
- In this notebook, we perform a data-driven analysis of flag images, focusing on their aspect ratios.

### Background from the Blog Post
- According to the blog post *"Aspect Ratios of Flags"* by Xawnia Wagner (Aug 21, 2017):
- **General Observation:**  
    - Almost all national flags have a rectangular shape with aspect ratios (longer side divided by shorter side) ranging between 1.0 (square) and 2.0.
- **Historical Common Ratios:**  
- Certain aspect ratios appear more frequently:
    - **1:1 (1.0):** Vatican, Switzerland  
    - **1:2 (2.0):** Canada, Croatia, Cuba, Ireland, etc.  
    - **2:3 (1.5):** Netherlands, Romania, Morocco, Turkey, China, etc.  
    - **3:5 (1.666â€¦):** Bulgaria, Bahrain, Luxembourg, Bangladesh, etc.  
    - **5:8 (1.6):** Poland, Sweden, Argentina, etc.

- **Additional Insights:**  
    - The blog also discusses how these ratios are influenced by history and design conventions (e.g., British, French, German influence) and even touches on design principles like the golden ratio.
    - In the following sections, we will:
        - 1. Load and preprocess flag image data.
        - 2. Compute and visualize the distribution of flag aspect ratios.
        -  3. Validate the common ratios mentioned in the blog post.
         - 4. Identify new insights such as outliers and clustering patterns.


## 1. Data Loading and Preprocessing
- We assume that the flag images (in PNG format) are stored in a local folder called `flags_png/`.
- Each image is named by its two-letter country code (e.g., `US.png`, `IN.png`).
- We will:
    - Load each image.
    - Extract its dimensions (width and height).
    - Compute its aspect ratio as **width/height** (which should yield values between 1.0 and 2.0 for landscape flags).


```python

import os
from PIL import Image
import pandas as pd

# Define the directory where flag images are stored
flags_dir = "flags_png"

# List all PNG files in the directory
flag_files = [f for f in os.listdir(flags_dir) if f.endswith('.png')]

# Create an empty list to collect data
flag_data = []

for file in flag_files:
    file_path = os.path.join(flags_dir, file)
    try:
        with Image.open(file_path) as img:
            width, height = img.size
            # Ensure width is the larger dimension (landscape orientation)
            if width < height:
                width, height = height, width  # swap if necessary
            aspect_ratio = width / height
            country = file.split('.')[0].upper()
            flag_data.append({
                "Country": country,
                "Width": width,
                "Height": height,
                "Aspect_Ratio": aspect_ratio
            })
    except Exception as e:
        print(f"Error processing {file}: {e}")

# Create a DataFrame from the collected data
df_flags = pd.DataFrame(flag_data)
print("First few rows of flag data:")
display(df_flags.head())
```

## 2. Exploratory Analysis of Aspect Ratios
- **We now explore the distribution of flag aspect ratios. We will:**
    - Compute summary statistics (mean, median, min, and max).
    - Visualize the distribution using a histogram.


```python

import matplotlib.pyplot as plt
import seaborn as sns

# Summary statistics
mean_ar = df_flags['Aspect_Ratio'].mean()
median_ar = df_flags['Aspect_Ratio'].median()
min_ar = df_flags['Aspect_Ratio'].min()
max_ar = df_flags['Aspect_Ratio'].max()

print(f"Mean Aspect Ratio: {mean_ar:.2f}")
print(f"Median Aspect Ratio: {median_ar:.2f}")
print(f"Min Aspect Ratio: {min_ar:.2f}")
print(f"Max Aspect Ratio: {max_ar:.2f}")

# Histogram with KDE
plt.figure(figsize=(10,6))
sns.histplot(df_flags['Aspect_Ratio'], bins=20, kde=True, color='skyblue')
plt.xlabel("Aspect Ratio (Width/Height)")
plt.ylabel("Frequency")
plt.title("Distribution of Flag Aspect Ratios")
plt.axvline(mean_ar, color='red', linestyle='--', label=f'Mean: {mean_ar:.2f}')
plt.axvline(median_ar, color='green', linestyle='--', label=f'Median: {median_ar:.2f}')
plt.legend()
plt.show()
```

## 3. Comparison with Blog Post Observations
- The blog post highlights several common aspect ratios. For our analysis, we consider the following:

| Flag Ratio Notation | Implied Value (Longer/Shorter) | Examples (from blog post)        |
|---------------------|--------------------------------|----------------------------------|
| 1:1                 | 1.0                            | Vatican, Switzerland             |
| 1:2                 | 2.0                            | Canada, Croatia, Cuba, Ireland   |
| 2:3                 | 1.5                            | Netherlands, Romania, Morocco, etc. |
| 3:5                 | ~1.67                          | Bulgaria, Bahrain, Luxembourg, etc. |
| 5:8                 | 1.6                            | Poland, Sweden, Argentina         |
    
- **We'll highlight flags with aspect ratios close to these values using a tolerance.**

```python

# Define common aspect ratios based on the blog post (values are longer_side/shorter_side)
common_ratios = {
    "1:1": 1.0,
    "1:2": 2.0,
    "2:3": 1.5,
    "3:5": 1.6667,
    "5:8": 1.6
}

tolerance = 0.05  # Acceptable deviation from common ratio

for label, value in common_ratios.items():
    nearby_flags = df_flags[(df_flags['Aspect_Ratio'] >= (value - tolerance)) &
                            (df_flags['Aspect_Ratio'] <= (value + tolerance))]
    print(f"\nFlags with aspect ratio near {label} ({value:.2f}):")
    if nearby_flags.empty:
        print("  None found.")
    else:
        display(nearby_flags[['Country', 'Aspect_Ratio']])
```

## 4. New Insights and Additional Observations
- **Beyond validating the blog post findings, our analysis reveals additional insights:**
    - **Variation Range:**  
        - The aspect ratios in our dataset range from **{min_ar:.2f}** to **{max_ar:.2f}**.
    - **Outlier Detection:**  
        -  We identify potential outliers using the Interquartile Range (IQR) method. This can highlight flags that deviate significantly from the typical ratios.
    - **Clustering Patterns:**  
        - A box plot helps to visually summarize the central tendency and spread of the aspect ratios.

#### Let's create a box plot and list outliers.


```python

# Box plot of aspect ratios
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_flags['Aspect_Ratio'], color='lightgreen')
plt.xlabel("Aspect Ratio (Width/Height)")
plt.title("Box Plot of Flag Aspect Ratios")
plt.show()

# Calculate IQR and define bounds for outliers
Q1 = df_flags['Aspect_Ratio'].quantile(0.25)
Q3 = df_flags['Aspect_Ratio'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_flags[(df_flags['Aspect_Ratio'] < lower_bound) | (df_flags['Aspect_Ratio'] > upper_bound)]
print("Flags identified as outliers based on aspect ratio:")
display(outliers[['Country', 'Aspect_Ratio']])
```

## 5. Standard Image Analysis Techniques on Flag Data

In addition to analyzing flag aspect ratios, we further explore the visual characteristics of flag images using standard image analysis techniques. These techniques include:

- **Pixel Intensity Histogram:**  
  Visualizes the distribution of grayscale pixel intensities, indicating the overall brightness and contrast of the flag.

- **Color Histograms:**  
  Analyzes the distribution of color intensities for each color channel (Red, Green, Blue) to understand the dominant colors.

- **Edge Detection:**  
  Applies edge detection (using Canny) to reveal structural features and design elements of the flag.

Below is the complete code example that performs these analyses on a sample flag image:


```python

import cv2  # OpenCV for image processing
import matplotlib.pyplot as plt
import numpy as np
import os


# Choose a sample flag image from flags_dir (e.g., the first file)
sample_flag_file = os.path.join(flags_dir, 'in.png')
print("Sample flag file:", sample_flag_file)

# Read the image using OpenCV
image = cv2.imread(sample_flag_file)
# Convert from BGR (OpenCV default) to RGB for proper display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

### 5.1 Grayscale Pixel Intensity Histogram

```python

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10, 6))
plt.hist(gray.ravel(), bins=256, range=[0, 256], color='gray')
plt.title("Grayscale Pixel Intensity Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

### 5.2 Color Histograms for Each Channel

```python

colors = ('r', 'g', 'b')
plt.figure(figsize=(10, 6))
for i, color in enumerate(colors):
    # Calculate the histogram for each color channel
    hist = cv2.calcHist([image_rgb], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)
    plt.xlim([0, 256])
plt.title("Color Histogram for Each Channel")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

### 5.3 Edge Detection using Canny

```python

edges = cv2.Canny(gray, 100, 200)
plt.figure(figsize=(10, 6))
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection (Canny)")
plt.axis("off")
plt.show()
```

## 5. Conclusion
- **Summary of Findings:**
    - Our analysis confirms that most national flags have aspect ratios between 1.0 and 2.0.
    - The histogram shows clear peaks near the common values:
    - Approximately **1.5** (consistent with a 2:3 ratio).
    - A cluster near **2.0** (consistent with a 1:2 ratio).
    - Additional clusters are observed near **1.6 - 1.67** (consistent with 5:8 and 3:5 ratios).
- **Outliers**: Some flags fall outside the typical IQR range, suggesting unique design choices.
- **New Observations:**
- The spread of aspect ratios indicates that while there are historical and design conventions, there is still some flexibility in flag dimensions.
- The clustering patterns may reflect regional influences or the historical adoption of specific ratios.
- Further analysis could correlate these aspect ratios with other visual or historical features of the flags.
