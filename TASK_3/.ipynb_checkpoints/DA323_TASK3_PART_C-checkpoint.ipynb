# Textual Analysis of National Anthem Translations
- In this notebook, we perform a textual analysis of the English translations of national anthems.
- **Objectives:**
    - 1. **Data Loading:**  
        - Load the anthem translation data from a JSON file (e.g., `anthem_translations.json`) that contains the English translations of national anthems.
    
    - 2. **Text Preprocessing:**  
        - **Preprocess the text by:**
            - Converting all text to lowercase.
            - Removing punctuation.
            - Tokenizing the text into words.
            - Removing common English stop words.
    
    - 3. **Exploratory Analysis:**  
        - **Analyze the processed text to identify the most common words and visualize the results using:**
            - Frequency distribution (bar plot).
            - Word cloud.

### This approach is analogous to the visual analysis performed on flag images but applied to text data.


## 1. Data Loading
- We start by loading the anthem translation data from the JSON file.
- **Note:** Each entry in the JSON file corresponds to a country's anthem translation.


```python

import json

# Load the JSON file containing anthem translations
with open("anthem_translations.json", "r", encoding="utf-8") as f:
    anthem_translations = json.load(f)

# Print a sample entry to see a snippet of the translation
sample_country = list(anthem_translations.keys())[0]
print(f"Anthem translation for {sample_country} (first 500 characters):\n")
print(anthem_translations[sample_country][:500])
```

## 2. Text Preprocessing
- In this section, we preprocess the text to prepare it for analysis.
- **Steps:**

     - **Combine Text:**  
       Combine all anthem translations into one text corpus.
    
     - **Lowercase Conversion:**  
       Convert all text to lowercase to ensure consistency.
    
     - **Punctuation Removal:**  
       Remove punctuation and any non-alphabetic characters.
    
     - **Tokenization:**  
       Split the text into individual words (tokens).
    
     - **Stop Word Removal:**  
       Remove common English stop words that do not contribute much to the meaning.

#### We use Python's `re` module for regex-based cleaning and NLTK for tokenization and stop word removal.


```python

import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download required NLTK resources (if not already available)
nltk.download('punkt')
nltk.download('stopwords')

# Combine all anthem translations into a single corpus
corpus = " ".join(anthem_translations.values())

# Convert the corpus to lowercase
corpus = corpus.lower()

# Remove punctuation and numbers (retain only alphabets and whitespace)
corpus = re.sub(r'[^a-z\s]', '', corpus)

# Tokenize the cleaned text into words
tokens = word_tokenize(corpus)
print(f"Total tokens before stop word removal: {len(tokens)}")

# Remove stop words using NLTK's English stopwords list
stop_words = set(stopwords.words("english"))
filtered_tokens = [word for word in tokens if word not in stop_words]

print(f"Total tokens after stop word removal: {len(filtered_tokens)}")
```

## 3. Frequency Analysis

- Next, we perform a frequency analysis on the processed tokens to identify the most common words in the anthem translations.

 - We use Pythonâ€™s `Counter` from the `collections` module to compute word frequencies.


```python
# %% [code]
from collections import Counter

# Create a frequency distribution of the filtered tokens
word_counts = Counter(filtered_tokens)

# Extract the 20 most common words
common_words = word_counts.most_common(20)
print("20 Most common words (after stop word removal):")
for word, count in common_words:
    print(f"{word}: {count}")
```

### 3.1. Visualization of Word Frequencies
- We now visualize the top 20 most common words using a bar plot.


```python

import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data for plotting
words, counts = zip(*common_words)

plt.figure(figsize=(12, 6))
sns.barplot(x=list(words), y=list(counts), palette="viridis")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top 20 Most Common Words in Anthem Translations")
plt.xticks(rotation=45)
plt.show()
```

## 4. Word Cloud Visualization

- A word cloud offers a visual summary of the most frequent words in the anthem translations.
- We use the `WordCloud` library to generate the visualization.


```python

from wordcloud import WordCloud

# Create a single string from the filtered tokens
wordcloud_text = " ".join(filtered_tokens)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(wordcloud_text)

plt.figure(figsize=(14, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Anthem Translations")
plt.show()
```

## 5. N-gram Analysis
- In addition to individual word frequencies, common phrases can provide valuable insights.
- We compute:

     - **Bigrams:** Two-word combinations.
     - **Trigrams:** Three-word combinations.

#### We'll use NLTK's `ngrams` function to generate these combinations from our filtered tokens.


```python

from nltk.util import ngrams
from collections import Counter

def get_ngrams(tokens, n=2):
    """
    Generate n-grams from a list of tokens.
    Returns a Counter object with n-gram frequencies.
    """
    return Counter(ngrams(tokens, n))

# Compute bigrams and trigrams
bigrams = get_ngrams(filtered_tokens, n=2)
trigrams = get_ngrams(filtered_tokens, n=3)

# Extract the 10 most common bigrams and trigrams
common_bigrams = bigrams.most_common(10)
common_trigrams = trigrams.most_common(10)

print("Top 10 Bigrams:")
for gram, count in common_bigrams:
    print(" ".join(gram), ":", count)

print("\nTop 10 Trigrams:")
for gram, count in common_trigrams:
    print(" ".join(gram), ":", count)
```

### 5.1. Visualization of N-grams

- We now visualize the most common bigrams and trigrams using bar plots.


```python

def plot_ngrams(ngrams_common, title):
    grams = [" ".join(gram) for gram, count in ngrams_common]
    counts = [count for gram, count in ngrams_common]
    plt.figure(figsize=(12, 6))
    sns.barplot(x=grams, y=counts, palette="magma")
    plt.xlabel("N-grams")
    plt.ylabel("Frequency")
    plt.title(title)
    plt.xticks(rotation=45)
    plt.show()

plot_ngrams(common_bigrams, "Top 10 Most Common Bigrams in Anthem Translations")
plot_ngrams(common_trigrams, "Top 10 Most Common Trigrams in Anthem Translations")
```

## 6. Sentiment Analysis

- To assess the overall tone of the anthem translations, we use sentiment analysis.

-  We will:

     - Use NLTK's VADER (Valence Aware Dictionary and sEntiment Reasoner) for sentiment scoring.
     - Compute sentiment scores for each anthem.
     - Visualize the distribution of sentiment scores.

 **Note:** VADER is designed for social media texts but works well for general sentiment analysis.


```python

import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

# Download VADER lexicon if not already done
nltk.download('vader_lexicon')

# Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Compute sentiment scores for each anthem translation
sentiment_scores = {}
for country, text in anthem_translations.items():
    # Clean the text similar to previous steps
    text_clean = text.lower()
    text_clean = re.sub(r'[^a-z\s]', '', text_clean)
    sentiment_scores[country] = sia.polarity_scores(text_clean)

# Convert the sentiment scores to a DataFrame for easier analysis
df_sentiments = pd.DataFrame.from_dict(sentiment_scores, orient='index')
print("Sample sentiment scores:")
display(df_sentiments.head())
```

### 6.1. Visualization of Sentiment Analysis

- We now visualize the sentiment analysis results.

- We'll look at the distribution of the **compound** sentiment scores, which combine positive, negative, and neutral scores into one metric.


```python

plt.figure(figsize=(10,6))
sns.histplot(df_sentiments['compound'], bins=20, kde=True, color='salmon')
plt.xlabel("Compound Sentiment Score")
plt.ylabel("Frequency")
plt.title("Distribution of Compound Sentiment Scores for Anthem Translations")
plt.axvline(df_sentiments['compound'].mean(), color='blue', linestyle='--', label=f"Mean: {df_sentiments['compound'].mean():.2f}")
plt.legend()
plt.show()
```

## 7. Topic Modeling

 - Finally, we apply topic modeling to uncover underlying themes in the anthem translations.

 - We'll use scikit-learn's Latent Dirichlet Allocation (LDA) for this purpose.

- **Steps:**

     - 1. Use `CountVectorizer` to convert the corpus of anthem translations into a document-term matrix.
     - 2. Fit an LDA model to the matrix.
     - 3. Extract and display the top words for each topic.

#### For simplicity, we treat each anthem as a separate document.


```python

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Prepare the corpus: list of cleaned anthem texts (one per country)
documents = []
countries = []
for country, text in anthem_translations.items():
    # Simple cleaning: lowercasing and punctuation removal (similar to before)
    text_clean = text.lower()
    text_clean = re.sub(r'[^a-z\s]', '', text_clean)
    documents.append(text_clean)
    countries.append(country)

# Use CountVectorizer to create the document-term matrix
vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2)
dtm = vectorizer.fit_transform(documents)
print("Document-Term Matrix shape:", dtm.shape)

# Fit an LDA model with a predefined number of topics (e.g., 3 topics)
num_topics = 3
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(dtm)

# Function to display the top words for each topic
def display_topics(model, feature_names, num_top_words):
    for topic_idx, topic in enumerate(model.components_):
        message = f"Topic #{topic_idx + 1}: "
        message += " | ".join([feature_names[i] for i in topic.argsort()[:-num_top_words - 1:-1]])
        print(message)

num_top_words = 10
print("Top words per topic:")
display_topics(lda, vectorizer.get_feature_names_out(), num_top_words)
```
